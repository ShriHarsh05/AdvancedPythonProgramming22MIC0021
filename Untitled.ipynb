```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from yellowbrick.classifier import ClassificationReport

# Load the dataset
df = pd.read_csv('Crop_recommendation.csv')

# --- Exploratory Data Analysis ---

# Display the first few rows of the dataframe
print("--- First 5 rows of the dataset ---")
print(df.head())
print("\n" + "="*50 + "\n")

# Display summary statistics
print("--- Dataset Description ---")
print(df.describe())
print("\n" + "="*50 + "\n")


# Heatmap to check for null/missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cmap="coolwarm")
plt.title("Heatmap of Missing Values")
plt.show()

# Distribution of temperature and pH
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.distplot(df['temperature'], color="purple", bins=15, hist_kws={'alpha': 0.2})
plt.title("Temperature Distribution")

plt.subplot(1, 2, 2)
sns.distplot(df['ph'], color="green", bins=15, hist_kws={'alpha': 0.2})
plt.title("pH Distribution")
plt.show()

# Check if the dataset is balanced
plt.figure(figsize=(10, 8))
sns.countplot(y='label', data=df, palette="plasma_r")
plt.title("Crop Distribution")
plt.show()

# Pairplot to visualize relationships between features
# sns.pairplot(df, hue='label') # This can be computationally expensive
# plt.show()


# Jointplot for rainfall and humidity for specific conditions
sns.jointplot(x="rainfall", y="humidity", data=df[(df['temperature'] < 30) & (df['rainfall'] > 120)], hue="label")
plt.suptitle("Rainfall vs. Humidity (Temp < 30C, Rainfall > 120mm)", y=1.02)
plt.show()

# Jointplot for Potassium and Nitrogen
sns.jointplot(x="K", y="N", data=df[(df['N'] > 40) & (df['K'] > 40)], hue="label")
plt.suptitle("Potassium vs. Nitrogen (N > 40, K > 40)", y=1.02)
plt.show()

# Jointplot for Potassium and Humidity
sns.jointplot(x="K", y="humidity", data=df, hue='label', height=8, s=30, alpha=0.7)
plt.suptitle("Potassium vs. Humidity", y=1.02)
plt.show()


# Boxplot for pH by crop
plt.figure(figsize=(10, 8))
sns.boxplot(y='label', x='ph', data=df)
plt.title("pH by Crop")
plt.show()


# Boxplot for Phosphorous when rainfall is high
plt.figure(figsize=(10, 8))
sns.boxplot(y='label', x='P', data=df[df['rainfall'] > 150])
plt.title("Phosphorous Levels with High Rainfall (>150mm)")
plt.show()

# Lineplot for Potassium and Rainfall with low humidity
plt.figure(figsize=(12, 6))
sns.lineplot(data=df[(df['humidity'] < 65)], x="K", y="rainfall", hue="label")
plt.title("Potassium vs. Rainfall (Humidity < 65%)")
plt.show()


# --- Data Preprocessing and Model Training ---

# Convert labels to categorical codes
c = df.label.astype('category')
targets = dict(enumerate(c.cat.categories))
df['target'] = c.cat.codes

# Define features (X) and target (y)
y = df.target
X = df[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]


# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# Scale the features
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build and train the Random Forest Classifier
clf = RandomForestClassifier(max_depth=4, n_estimators=100, random_state=42).fit(X_train, y_train)

# Print accuracy scores
print('RF Accuracy on training set: {:.2f}'.format(clf.score(X_train, y_train)))
print('RF Accuracy on test set: {:.2f}'.format(clf.score(X_test, y_test)))
print("\n" + "="*50 + "\n")


# --- Model Evaluation ---

# Classification Report using Yellowbrick
classes = list(targets.values())
visualizer = ClassificationReport(clf, classes=classes, support=True, cmap="Blues")

visualizer.fit(X_train, y_train)  # Fit the visualizer and the model
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show()


# --- Crop Recommendation ---

# Input data for prediction
input_data = [[19, 72, 15, 28.83600962, 69.76112921, 6.890760124, 44.08562546]]

# Make a prediction
prediction = clf.predict(input_data)

# Get the recommended crop name
recommended_crop = targets[prediction[0]]

print(f"Recommendation for input {input_data}: {recommended_crop}")
```
