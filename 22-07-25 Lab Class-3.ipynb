# create a list of travel destination


```python
travel_dest = ["Hyderabad","Kashmir", "Goa", "Gujarat"]
#list of travel destination
print(travel_dest)
```

```python
print(len(travel_dest))
```

# Using the insert method -> used to add an element into list at a specific index

```python
travel_dest.insert(0, "Rajasthan")
print(travel_dest)
```

# Using the append method -> Used to add an element in the last index or value

```python
travel_dest.append("Chennai")
print(travel_dest)
```

# Using Extends Methods -> Used to merge 2 different list into 1

```python
#creating another list and appending it with the initial list
other_dest = ["Mumbai", "Goa", "USA"]
travel_dest.extend(other_dest)
print(travel_dest)
#We can use extend method for merging to lists into 1
```

# Using the pop method -> Used to delete the last element in the list

```python
travel_dest.pop()
print(travel_dest)
#USA is removed
```

# Using the remove() method -> Used to remove an element from the list with a specific value

```python
travel_dest.remove("Goa")
print(travel_dest)
# First instance of Goa is removed, right after Kashmir
```

# Using pop with index attribute -> Used to remove an element from the list with the specific index value

```python
travel_dest.pop(6)
print(travel_dest)
#Instance of Goa is removed
```

# Using delete function -> Used to remove an element from the list

```python
del travel_dest[0]
print(travel_dest)
```

# Using clear method -> Method is used to remove all elements from the list

```python
travel_dest.clear()
print(travel_dest)
```

# Create a platform for a Travel Recommendation Platform

```python
# Sample destination database with tags
dest = {
    "Goa": ["beach", "party", "nature"],
    "Kashmir": ["nature", "mountains", "snow"],
    "Jaipur": ["history", "culture", "desert"],
    "Kerala": ["backwaters", "nature", "beach"],
    "Hyderabad": ["history", "culture", "city"],
    "Manali": ["mountains", "snow", "nature"],
    "Andaman": ["beach", "island", "scuba"],
}
print(dest)
```

```python
# Ask user for their preferences
user_pref = input("Enter your travel preferences (comma separated): ").lower().split(',')
user_pref = [pref.strip() for pref in user_pref]
print(f'Enterd user preferences are: {user_pref}')
```

```python
# Ask user for previously visited places
visited_inp = input("Enter places you've already visited (comma separated): ").title().split(',')
visited = [place.strip() for place in visited_inp]
print(f'Enterd user visited places are: {visited}')
```

```python
# Recommendation logic
recommendations = []
for place, tags in dest.items():
    if place not in visited:
        if any(pref in tags for pref in user_pref):
            recommendations.append(place)
```

```python
# Output
if recommendations:
    print("\nRecommended places for you based on your preferences:")
    for place in recommendations:
        print("-", place)
else:
    print("\nSorry, no matching recommendations found based on your preferences.")
```
