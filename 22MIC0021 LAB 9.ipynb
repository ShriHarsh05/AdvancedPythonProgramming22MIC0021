## Multi-level and Multiple Inheritance in Python

```python
class Emp:
    no_of_emp = 0
    raise_amount = 1.04
#no_of_emp: A class variable to track the total number of employees created.
#raise_amount: A class variable for the standard raise percentage.
    def __init__(self,first,last,pay):
        self.last = last
        self.first = first
        self.pay = pay
        Emp.no_of_emp += 1
    
    def apply_raise(self):
        self.pay = int(self.pay * Emp.raise_amount)
        
```

```python
print(Emp.no_of_emp)
```

```python
e1 = Emp('Shriharsh','Kotecha',100000)
```

```python
print(Emp.no_of_emp)
```

```python
print(e1.pay)
e1.apply_raise()
print(e1.pay)
```

```python
e2 = Emp('S','k',200000)
```

## Single Inheritance

```python
class Parent:
    def greet(self):
        print('Hello from parent')
class Child(Parent):
    def __init__(self):
        self.greet()
        print("Hi from child")
    pass
obj1 = Child()
```

## Method Overriding and super()

```python
class Parent:
    def greet(self):
        print('Hello from parent')
class Child(Parent):
    def greet(self):
        super().greet()
        print('Hello from child')
obj1 = Child()
obj1.greet()
```

## Accessing Class Variables from a Child Class:

```python
class ChildEmp(Emp):
    def __init__(self):
        print(f"Amount of Raise per year is: {super().raise_amount}")
        print(f"No of Employees are :{super().no_of_emp}")
c1 = ChildEmp()
```

## multiple inheritence

```python
class A:
    def method_a(self):
        print("Method A")
class B:
    def method_b(self):
        print("Method B")
class C(A,B):
    pass
obj = C()
obj.method_a()
obj.method_b()
# class C` inherits from both `A` and `B`, gaining access to methods from both parent classes.
```

## multi-level inheritence

```python
class D:
    def method_a(self):
        print("Method A")
class E(D):
    def method_b(self):
        print("Method B")
class F(E):
    pass
obj = F()
obj.method_a()
obj.method_b()
```

## Practical Inheritance Example

```python
class Contact():
    all_contact = []
    def __init__(self,name,email):
        self.name= name
        self.email = email
        Contact.all_contact.append(self)
```

```python
class Supplier(Contact):
    def order(self,order):
        print("Send ""{} order to {}".format(order, self.name))
```

```python
c = Contact("Some Body","somebody@gmail.com")
s = Supplier("one body","onebody@gmail.com")
```

```python
print(c.name, c.email, s.name, s.email)
```
