# Functions with multiple arguments

```python
def get_full_name(first, last, middle=''):
    if middle:
        return f"{first} {middle} {last}"
    else:
        return f"{first} {last}"
print(get_full_name('Shriharsh','Kotecha'))
print(get_full_name('Shriharsh','Kotecha','S'))
```

# Passing a List

```python
def greet_users(names):
    for name in names:
        print(f"Hello, {name}!")
n = ['Shriharsh','Mayank']
greet_users(n)
```

# Modifying a List Inside the Function

```python
def modify_list(my_list):
    print("Inside function (before):", my_list)
    my_list.append("new_item")
    print("Inside function (after):", my_list)

items = ["item1", "item2"]
print("Before function call:", items)

modify_list(items)

print("After function call:", items)
```

# Protecting a List from Updates

```python
def modify_list(my_list):
    print("Inside function (before):", my_list)
    my_list.append("new_item")
    print("Inside function (after):", my_list)

items = ["item1", "item2"]
print("Before function call:", items)

modify_list(items[:])  # Pass a copy

print("After function call:", items)
```

# Passing an Arbitrary Number of Arguments

```python
def make_pizza(*toppings):
    for topping in toppings:
        print(f"- {topping}")

# Function call
make_pizza('pepperoni', 'pineapple', 'onions')
```

# Mixing Positional and Arbitrary Arguments

```python
def make_pizza(size, *toppings):
    print(f"Making a {size}-inch pizza with:")
    for topping in toppings:
        print(f"- {topping}")

# Function call
make_pizza(12, 'extra cheese', 'tomato', 'onions')
```

```python
def build_profile(first, last, **info):
    profile = {'first': first, 'last': last}
    profile.update(info)
    return profile

# Function call
user_profile = build_profile('Shriharsh', 'Kotecha', location='India', field='Computer Science', year='2027')
print(user_profile)
```
