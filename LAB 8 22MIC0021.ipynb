# Simple multiply by 2 function

```python
def double(x):
    return x*2
print(double(2))
my_lambda = lambda x: x*2
print(my_lambda(2))
```

# Addition using Lambda function vs normal defination function 

```python
def add(x,y):
    return x+y
print(add(5,6))

add_lambda = lambda x,y: x+y
print('Addition using Lambda: ',add_lambda(5,5))
```

# Conditional Logic Lambda function

```python
def mx(x,y):
    if x>y:
        return x
    else:
        return y

compare_mx = lambda x,y : x if x>y else y
print('Using normal function',mx(7,8))
print('Using lambda function',compare_mx(5,6))
```

```python
names = ['Ema', 'Tom', 'hinton']
looper = names.__iter__()
print(looper)
while True:
    try:
        print(looper.__next__())
    except StopIteration:
        print("End of names list reached.")  # Handle the StopIteration exception
        break
```

```python
names = ['Ema', 'Tom', 'hinton']
looper = iter(names)
print(looper)
while True:
    try:
        print(next(looper))
    except StopIteration:
        print("End of names list reached.")  # Handle the StopIteration exception
        break
```

# Using map() function

```python
def square(n):
    return n*n
m=[1,2,3,4,5]
print(list(map(square,[1,2,3,4,5])))

def square(n):
    return n**2
m=[1,2,3,4,5]
print(list(map(square,m)))
```

```python
m=[1,2,3,4,5]
print(list(map((lambda x: x * x),[1,2,3,4,5])))
```

```python
print(list(filter(lambda x: x>3,m)))
# each element is individually passed and then the condition is checked and if passed then the element is added to the list
```

```python
def toL(stri):
    return stri.lower()
print(toL('Shriharsh'))

names = ['SHRIHARSH','S','KOTECHA']
print(list(map(toL,names)))
print(list(map(lambda x:x.lower(),names)))
```

# Using reduce() function

```python
from functools import reduce
numbers = [1,2,3,4,5]
print(reduce(lambda x,y: x+y, numbers))
# This function is cumulatively adding values of the list and storing in variable x
```

```python
quote = ['Now,','this','is','not','the','end.','It','is','not','even','the','begining','of','the','end.','But','it','is,','perhaps,','the','end','of','the','beginning']
print(reduce(lambda x,y: x+" "+y, quote))
```
