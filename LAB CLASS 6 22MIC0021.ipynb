# Sequence Unpacking

```python
# Unpacking elements from a list into individual variables
odds = [1,3,5,7]
a,b,c,d = odds
print(a)
print(b)
print(c)
print(d)
```

# Sequence Unpacking with * (Extended Unpacking)

```python
a,b,*c = odds
print(a) #OUTPUT 1
print(b) #OUTPUT 3
print(c) #OUTPUT [5,7]
```

# Collecting the middle element

```python
a,*b,c = [3,5,7,9]
print(a)
print(b)
print(c)
# Here, a gets the first item, c gets the last, and *b gets the middle items.
```

# Using * to Unpack Arguments into a Function

```python
odds = [1,3,5]
def a(x,y,z):
    return (x+y+z)

# Passing a list using unpacking
print(a(*odds))
```

# other of function call method with the same output

```python
def a(x,y,z):
    return (x+y+z)
print(a(1,3,5))
```

# Function for finding sum of a list with 'n' number of items

```python
def a(*var):
    sum = 0
    for n in var:
        sum += n
    return sum
print(a(1,3,5))
```

# Function for finding sum of a list containing 'n' terms and calling it dynamically

```python
def a(*var):
    sum = 0
    for n in var:
        sum += n
    return sum
print(a(*odds))
```

# Using Sequence Unpacking with dictionary

```python
def show_info(**info):
    for k,v in info.items():
        print(f"{k}:{v}")
show_info(Code="CSI3007",Name="Python",Venue="SJT318")
```
