```python
import pyaudio
print(pyaudio.__version__)
```

# 16 bit integer format

```python
FORMAT = pyaudio.paInt16
```

# no of audio channels 1 as in mono, 2 as in stereo


```python
CHANNELS = 2
```

```python
# Rate as in 44100 samples per second
RATE = 44100
CHUNK = 1024  # Frames per buffer
# Initializes PyAudio, preparing the interface for audio streaming.
p = pyaudio.PyAudio()

# input stream to record 
stream_in = p.open(format=FORMAT,
                   channels=CHANNELS,
                   rate=RATE,
                   input=True,
                   frames_per_buffer=CHUNK)

# output stream to send audio (speaker)
stream_out = p.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    output=True,
                    frames_per_buffer=CHUNK)

print("Start speaking")

try:
    while True:
        data = stream_in.read(CHUNK)
        stream_out.write(data)
except KeyboardInterrupt:
    print("\nStopped.")

stream_in.stop_stream()
stream_in.close()
stream_out.stop_stream()
stream_out.close()
p.terminate()
```
